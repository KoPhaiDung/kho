```python
#W5A1
def Max_Of_Two(m,n):
    return max(m,n)
a,b=map(int,input().split())
print(Max_Of_Two(a,b))
```

```python
#W5A3
def isprime(n):
    for i in range(2,int(n**(1/2)+1)):
        if n%i==0:
            return False
    return True
a=int(input())
if isprime(a)==False:
    print(False)
else:print(True)
```

```python
#W5A4
def isperfect(n):
    sum=0
    for i in range(2,int(n**(1/2))+1):
        if n%i==0:
            sum+=i
            sum+=n/i
    if sum+1==n:return True
    return False
a=int(input())
if isperfect(a):
    print(True)
else:print(False)
```

```python
#W5A5
def check_list(o,k):
    count=0
    for i in o:
        count+=1
        if i==k:
            return count
    return -1
l=list(map(int,input().split()))
k=int(input())
print(check_list(l,k))
```

```python
#W5A6
def giai_thua(n):
    if(n==0):return 1
    if(n==1):return 1
    return n*giai_thua(n-1)
a=int(input())
print(giai_thua(a))
```

```python
#W5A8
def chuyen_nhi_phan(n):
    xau=''
    while n!=1:
        if n%2==0:
            xau+='0'
            n/=2
        else:
            xau+='1'
            n//=2
    return len(xau)
a,b=map(int,input().split())
print(abs(chuyen_nhi_phan(a)-chuyen_nhi_phan(b)))
```
