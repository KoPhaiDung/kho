```python
#W6A9
def nhap_dict(m):
    dic={}
    m=m[1:len(m)-1:]
    l=list(map(str,m.split(",")))
    for i in l:
        key,value=i.split(":")
        key=key[1:len(key)-1:]
        if key[0]=="'": key=key[1::]
        value=int(value)
        dic[key]=value
    return dic
a=input()
b=input()
dica=nhap_dict(a)
dicb=nhap_dict(b)
if len(dica)>len(dicb):dica,dicb=dicb,dica
dict={}
for i in dica:
    for j in dicb:
        if i == j :
            dica[i]+=dicb[j]
        elif j not in dica:
            dict[j]=dicb[j]
for i in dict:
    dica[i]=dict[i]
print(dica)
```

```python
#W6A10
def buble_sort(l):
    for i in range(0,len(l)):
        is_swap=False
        for j in range(0,len(l)-1):
            if l[j][0]>l[j+1][0]:
                l[j],l[j+1]=l[j+1],l[j]
                is_swap=True
        if is_swap== False:
            break
    return l
l=list(map(int,input().split()))
#l.sort()
k=int(input())
fix_l={}
d=0
res_l=[]
for i in l:
    if i not in fix_l.values():
        fix_l[d]=i
        d+=1
for i in range(0,len(fix_l)//2):
    if k-fix_l[i] in fix_l.values() :
        n=(fix_l[i],k-fix_l[i])
        res_l.append(n)
res_l=buble_sort(res_l)
print(res_l)
```
