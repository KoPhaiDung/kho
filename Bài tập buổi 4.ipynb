```python
#W4A7
import  math as m
def sang(n):
    ktra=[True]*100000
    j=0
    r=int(m.sqrt(n))
    for i in range(2,r+1):
        if ktra[i]==False:continue
        j=i
        while(j<r):
            j+=i
            ktra[j]=False
    for i in range(r,1,-1):
        if ktra[i]==True:
            if n%i==0:
                return i
n=int(input())
print(sang(n))
```

```python
#W4A8
def dao_nguoc(a):
    a=str(a)
    dao=''
    for i in range(len(a),0,-1):
        dao+=a[i-1]
    dao=int(dao)
    return dao
n=int(input())
d=0
while(n!=dao_nguoc(n)):
    d+=1
    n+=dao_nguoc(n)
print(d,n)
```

```python
#W4A10
def collatz(n):
    cnt =0
    while n!=1:
        if n%2==1:
            n=3*n+1
            cnt+=1
        else:
            n//=2
            cnt+=1
    return cnt+1
n = int(input())
x=0
ans=0
for i in range(1, n+1):
    temp = collatz(i)
    if temp>ans:
        ans = temp
        x = i
print(x, ans)
```
